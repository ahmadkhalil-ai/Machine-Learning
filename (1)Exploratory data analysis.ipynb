# Visualization

```python
fig = px.histogram(df,
    x='Size',
    marginal='box',
    nbins=47,
    title='Size of the banana')
fig.update_layout(bargap=0.1)
fig.show()
```

```python
fig = px.histogram(df,
                   x = 'Weight',
                   marginal='box',
                   color_discrete_sequence=['red'],
                   title='Weight of the banana')
fig.update_layout(bargap=0.1)
fig.show()
```

```python
fig = px.histogram(df,
                  x = "Ripeness",
                  marginal = 'box',
                  color='Quality',
                  nbins = 20,
                  color_discrete_sequence=['red','green'],
                  title = 'Ripeness of banana')
fig.update_layout(bargap=0.1)
fig.show()
```

```python
fig = px.scatter(df,
                  x = "Ripeness",
                  y = 'Sweetness',
                  color='Quality',
                  opacity=0.8,
                  hover_data=['Softness'],
                  title = 'Ripness vs Sweetness')
fig.update_traces(marker_size=5)
fig.show()
```

```python
fig = px.scatter(df,
                  x = "Softness",
                  y = 'Acidity',
                  color='Quality',
                  opacity=0.6,
                  hover_data=['Softness'],
                  title = 'Softness vs Acidity')
fig.update_traces(marker_size=5)
fig.show()
```
